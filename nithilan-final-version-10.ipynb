Hyperparameter Tuning:

```python
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Define the hyperparameter combinations to test
n_estimators = [100, 200, 300]
max_depth = [10, 15, 20]
min_samples_split = [2, 5, 10]

best_score = float('inf')
best_params = {}

for n_est in n_estimators:
    for depth in max_depth:
        for min_split in min_samples_split:
            rf = tfdf.keras.RandomForestModel(
                task=tfdf.keras.Task.REGRESSION,
                winner_take_all=True,
                categorical_algorithm='RANDOM',
                split_axis='SPARSE_OBLIQUE',
                sparse_oblique_normalization='MIN_MAX',
                sparse_oblique_num_projections_exponent=1.0,
                num_trees=n_est,
                max_depth=depth,
                min_examples=min_split
            )
            
            # Convert and split the dataset
            train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train_ds_pd, label='SalePrice', task=tfdf.keras.Task.REGRESSION)
            valid_ds = tfdf.keras.pd_dataframe_to_tf_dataset(valid_ds_pd, label='SalePrice', task=tfdf.keras.Task.REGRESSION)
            
            # Train the model
            rf.fit(x=train_ds)
            
            # Make predictions on the validation set
            predictions = rf.predict(valid_ds)
            
            # Calculate the RMSE
            rmse = np.sqrt(mean_squared_error(valid_ds_pd['SalePrice'], predictions))
            
            if rmse < best_score:
                best_score = rmse
                best_params = {'n_estimators': n_est, 'max_depth': depth, 'min_samples_split': min_split}

print("Best hyperparameters:", best_params)
print("Best RMSE:", best_score)
```

Semi-Supervised Learning:




```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.semi_supervised import SelfTrainingClassifier

# Load a sample dataset (Iris dataset in this example)
iris = load_iris()
X, y = iris.data, iris.target

# Split the data into labeled and unlabeled parts
X_labeled, X_unlabeled, y_labeled, _ = train_test_split(X, y, test_size=0.7, random_state=42)

# Create a base model (RandomForestClassifier in this example)
base_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the base model on the labeled data
base_model.fit(X_labeled, y_labeled)

# Self-training with the base model
self_training_model = SelfTrainingClassifier(base_model)

# Train the self-training model on the labeled and unlabeled data
self_training_model.fit(X, y)

# Make predictions with the self-training model
predictions = self_training_model.predict(X)

# Evaluate the model's performance (for demonstration purposes)
accuracy = sum(predictions == y) / len(y)
print(f"Accuracy of Self-training Model: {accuracy:.2f}")
```

```python
import shap
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Generate some sample data (replace this with your dataset)
data = pd.DataFrame(np.random.rand(100, 5), columns=['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5'])
target = 3 * data['Feature1'] + 2 * data['Feature2'] - data['Feature3'] + np.random.randn(100) * 0.1

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)

# Train a simple model (you should replace this with your regression model)
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Explain the model using SHAP
explainer = shap.Explainer(model)
shap_values = explainer.shap_values(X_test)

# Create a summary plot
shap.summary_plot(shap_values, X_test)

# Evaluate the model
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```
